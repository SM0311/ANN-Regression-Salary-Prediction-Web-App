```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder,OneHotEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
#Preprocess the data
data = data.drop(['RowNumber','CustomerId','Surname'], axis=1)
data.head()
```

```python
## Encode Categorical Variable 
lable_regression_encoder_gender = LabelEncoder()
data['Gender'] = lable_regression_encoder_gender.fit_transform(data['Gender'])
```

```python
data['Gender']
```

```python
data.head()
```

```python
# One-hot encoder 'Geography'
onehot_regression_encoder_geo = OneHotEncoder()
geo_encoded = onehot_regression_encoder_geo.fit_transform(data[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded,columns=onehot_regression_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
data.head()
```

```python
data = data.drop('Geography',axis=1)
data
```

```python
data = pd.concat([data,geo_encoded_df],axis=1)
data
```

```python
# Split the data into train and test split 
X = data.drop('EstimatedSalary',axis=1)
y = data['EstimatedSalary']
```

```python
X
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)


# Scale the features
## Scale these features 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y
```

```python
# Save the encoders and scaler

with open('lable_encoder_gender','wb') as file:
    pickle.dump(lable_regression_encoder_gender, file)

with open('onehot_encoder_geo','wb') as file:
    pickle.dump(onehot_regression_encoder_geo, file)

with open('scaler.pkl','wb') as file:
    pickle.dump(scaler, file)
```

#### ANN Regression Problem Statement 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Build the model 
model = Sequential([
    Dense(64, activation =  'relu',input_shape = (X_train.shape[1],)),
    Dense(32, activation = 'relu'),
    Dense(1) #output layer
])

# Complie the model 
model.compile(optimizer = 'adam', loss = 'mean_absolute_error', metrics = ['mae'])

# Summary 
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime

## Set up the Tensor Board
log_dir = "regressionlogs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq= 1)
```

```python
## Set up Eraly Stopping 
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
## Train the model 
history = model.fit(
    X_train,y_train,
    validation_data =  (X_test,y_test),
    epochs = 100,
    callbacks = [early_stopping_callback,tensorboard_callback]
)
```

```python
## Evaluate model on the test data
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test MAE: {test_mae:.2f}')
```
